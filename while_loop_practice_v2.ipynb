Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
a=1

while a<=20:
  print(a)
  a= a+1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
a=1
n= int(input("enter input:" ))
sum=0
while a<=n:
  sum= sum+a
  a= a+1
print(f"sum of numbers from 1 to {n} is {sum}")

```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
n= 2345
rev= 0

while n>0:
  rev= rev*10+n%10
  n= n//10
print(f"reversed number is {rev}")
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
a= 123
count=0

if a==0:
  count=1
else:
  while a>0:
   a=a//10
   count+=1
print(count)




```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
n= 111
rev= 0

while n>0:
  rev= rev*10+n%10
  n= n//10
print(f"reversed number is {rev}")
```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
correct_num= 123
a= int(input("enter number: "))
while a!= correct_num:
   print("wrong")
   a= int(input("enter number: "))
print("correct guess")
```
