## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
n= int(input("enter: "))
count_even=0
count_odd=0

for i in range(1,n):
  if i%2==0:
    count_even=count_even+1
  else:
    count_odd=count_odd+1

print("even number between 1 to 40 is: ",count_even)
print("odd number between 1 to 40 is: ",count_odd)

```

Loop from 2 to N-1 and use `break` if divisible.

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
n = int(input("Enter n: "))

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:
1
12
123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
N = int(input("Enter N: "))

for i in range(1, N + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()


```
